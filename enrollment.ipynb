```python
df = pd.read_csv("enrollment2024.csv")
```

```python
import plotly.express as px
import pandas as pd

df_long = df.melt(id_vars=['School Name', 'Lat', 'Long','Difference'], 
                  value_vars=['2021', '2022', '2023', '2024'], 
                  var_name='Year', value_name='Enrollment')

df_long = df_long.dropna(subset=['Enrollment'])

fig = px.scatter_mapbox(df_long, 
                        lat="Lat", 
                        lon="Long", 
                        size="Enrollment", 
                        color="Difference",
                        hover_name="School Name", 
                        hover_data=["Year", "Enrollment"], 
                        animation_frame="Year", 
                        title="School Enrollment from 2021 to 2024",
                        mapbox_style="carto-positron", 
                        size_max=50, 
                        zoom=10)
#margin={"r":0,"t":0,"l":0,"b":0},
fig.update_layout(width=1200,
                 height=800)

fig.show()
```
